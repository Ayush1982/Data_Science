# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

ArithmeticError class is a base class for exceptions that occur during arithmetic operations. Two commonly used exception classes that inherit from ArithmeticError are ZeroDivisionError and OverflowError. Let's explain these two errors with examples:

1)ZeroDivisionError:
 
This error occurs when you try to divide a number by zero, which is mathematically undefined.

```python
dividend = 10
divisor = 0

try:
    result = dividend / divisor
    logging.info("all good")
    
except ZeroDivisionError as e:
    print(f"Error: {e}")
    access(f"error: {e}")
    logging.error("zeroDIV error occor")
```

2)OverflowError:

This error occurs when an arithmetic operation exceeds the limits of the data type being used, typically in situations where the result is too large to be represented.

```python
max_int = 2**31 - 1  
result = max_int + 1

try:
    result = max_int + 1
    logging.info("all good")
except OverflowError as e:
    print(f"Error: {e}")
    access(f"Error: {e}")
    logging.error("overflowError")
```

# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

LookupError class in Python is a base class for exceptions that occur when you try to access an element in a sequence using an invalid or non-existent key or index. It is not meant to be raised directly but serves as a base class for more specific lookup-related exception classes

```python
#KeyError:
my_dict = {'apple': 2, 'banana': 3, 'cherry': 5}

try:
    value = my_dict['grape']
    logging.info("currintly in try block")

except KeyError as e:
    print(f"Error: {e}")
    access(f"Error: {e}")
    logging.error(f"error : {e}")

#IndexError:
my_list = [10, 20, 30, 40, 50]

try:
    value = my_list[10]
    logging.info("all good")
except IndexError as e:
    print(f"Error: {e}")
    access(f"Error: {e}")
    logging.error(f"error : {e}")
```

# Q5. Explain ImportError. What is ModuleNotFoundError?

1)ImportError:

ImportError is a base class for exceptions that occur when there is an issue while importing a module. It is a more general exception, and other, more specific import-related exceptions, such as ModuleNotFoundError, can be derived from it.

2)ModuleNotFoundError:

ModuleNotFoundError is a specific subclass of ImportError . It is raised when Python cannot find the module you are trying to import. This error is more precise in indicating that the module you want to use is missing.

```python
#1)ImportError:
try:
    import non_existent_module
    logging.info("all good")
except ImportError as e:
    print(f"Error: {e}")
    access(f"Error: {e}")
    logging.error(f"error : {e}")

#2)ModuleNotFoundError:
try:
    import non_existent_module
    logging.info("all good")
except ModuleNotFoundError as e:
    print(f"Error: {e}")
    access(f"Error: {e}")
    logging.error(f"error : {e}")
```
