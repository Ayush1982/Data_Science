# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a data visualization library in Python that is based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1. **Scatter Plot:**
   - **Use:** Shows the relationship between two continuous variables. Each point represents an observation in the dataset.
   - **Seaborn Function:** `seaborn.scatterplot()`

2. **Line Plot:**
   - **Use:** Displays the trend of a continuous variable over a specific time or sequence. It's useful for time series data.
   - **Seaborn Function:** `seaborn.lineplot()`

3. **Bar Plot:**
   - **Use:** Compares categories and shows the distribution of a categorical variable. It's useful for comparing values across different groups.
   - **Seaborn Function:** `seaborn.barplot()`

4. **Histogram:**
   - **Use:** Displays the distribution of a single variable and shows the frequency of each bin in a continuous dataset.
   - **Seaborn Function:** `seaborn.histplot()`

5. **Box Plot (Box-and-Whisker Plot):**
   - **Use:** Visualizes the distribution of a dataset and its central tendency. It is useful for identifying outliers and understanding the spread of the data.
   - **Seaborn Function:** `seaborn.boxplot()`

Each of these plots can be customized further in Seaborn to enhance their visual appeal and convey specific insights from the data.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn
import matplotlib.pyplot as plt

fmri_data = seaborn.load_dataset("fmri")

plt.figure(figsize=(10, 6))
seaborn.lineplot(x="timepoint", y="signal", data=fmri_data)

plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("1")

plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

Note:  pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

sns.boxplot(x='pclass', y='age', data=titanic_data, ax=axes[0])
axes[0].set_title('Age Across Passenger Classes')
sns.boxplot(x='pclass', y='fare', data=titanic_data, ax=axes[1])
axes[1].set_title('Fare Across Passenger Classes')

plt.tight_layout()
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds_data = sns.load_dataset("diamonds")

plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut')

plt.xlabel('Price')
plt.ylabel('Cut')
plt.title('Histogram of Diamond Price')

plt.show()
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = sns.load_dataset("iris")
sns.pairplot(iris_data, hue='species', markers=['o', 's', 'D'])
plt.show()
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")

flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')
flights_pivot = flights_pivot.astype(int)
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')

plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
