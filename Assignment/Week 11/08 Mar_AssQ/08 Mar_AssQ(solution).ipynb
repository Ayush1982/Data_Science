#  Q1. What is the Probability density function?

A Probability Density Function is a mathematical function that tells about continuous random variable taking on a particular value within a given range. In simpler terms, it tells us how probable different outcomes are in a continuous distribution. For example, in a normal distribution, the PDF shows us the probabilities of different values occurring along the bell curve. It's like a blueprint that helps us understand the distribution of probabilities across a range of possible outcomes.

# Q2. What are the types of Probability distribution?

1. **Discrete Probability Distributions**: These distributions are defined for discrete random variables, which take on a finite or countably infinite number of distinct values. Examples include the Bernoulli distribution, binomial distribution, Poisson distribution, and geometric distribution.

2. **Continuous Probability Distributions**: Continuous distributions are defined for continuous random variables, which can take on any value within a certain range. Examples include the uniform distribution, normal (Gaussian) distribution, exponential distribution, and gamma distribution.

3. **Multinomial Distribution**: This distribution generalizes the binomial distribution to more than two categories. It describes the probabilities of observing counts among multiple categories.

4. **Poisson Distribution**: This distribution models the number of events occurring in a fixed interval of time or space, given a known average rate of occurrence.

#  Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
import scipy.stats as stats

mean = 0
std_dev = 1
x = 1

pdf_value = stats.norm.pdf(x, loc=mean, scale=std_dev)
print(f"PDF at x = {x}: {pdf_value}")
```

# Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

**Binomial distribution**: The binomial distribution is defined for discrete random variables, representing the number of successes in a fixed number of independent Bernoulli trials.Each trial in a binomial experiment can result in one of two outcomes: success or failure.The number of trials (n) is fixed and known in advance.Each trial is independent of the others, meaning the outcome of one trial does not affect the outcomes of the other trials.The probability of success (p) remains constant from trial to trial.The probability mass function (PMF) of the binomial distribution gives the probability of observing a specific number of successes in the fixed number of trials.

Two examples :

1. **Coin Flipping**: Suppose you flip a fair coin (where the probability of getting heads is 0.5) 10 times. The number of heads obtained in these 10 flips follows a binomial distribution because each flip is independent, with a constant probability of success (getting heads).

2. **Drug Testing**: Consider a clinical trial where a new drug is tested on a group of patients. Each patient either responds positively (success) or negatively (failure) to the drug, and their responses are independent of each other. The binomial distribution can be used to model the number of patients who respond positively out of a fixed number of patients tested.

#  Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000  
p = 0.4  

random_sample = np.random.binomial(n, p, size=1000)

plt.hist(random_sample, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Binomial Distribution (n=1000, p=0.4)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#  Q7. How Binomial distribution different from Poisson distribution?

1. **Binomial Distribution**:

   - The probability of success (p) remains constant from trial to trial.
   - The number of trials (n) is fixed and known in advance.
   - Each trial can result in one of two outcomes (success or failure).
   - Used when there is a fixed number of independent trials with two possible outcomes, such as the number of successes in a fixed number of coin flips.
   - Parameterized by the number of trials (n) and the probability of success (p).

2. **Poisson Distribution**:

   - The number of trials is not fixed; it is theoretically infinite.
   - Each trial can result in multiple outcomes (e.g., events occurring within a fixed interval of time or space).
   - Used to model the number of events occurring in a fixed interval of time or space when the events happen independently of the time since the last event, such as the number of phone calls received at a call center in a fixed time period.
   - Parameterized by the average rate of success (λ).
   - The probability of success (λ) is the average rate at which events occur in a fixed interval of time or space.


#  Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

mean = 5
sample_size = 1000

sample_mean  = np.random.poisson(mean, size=sample_size).mean()
sample_variance = np.random.poisson(mean, size=sample_size).var()

print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

#  Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

1. **Binomial Distribution**:
   - Mean (μ) = n * p, where n is the number of trials and p is the probability of success in each trial.
   - Variance (σ^2) = n * p * (1 - p)

   In the binomial distribution, the variance is directly related to the mean through the formula for variance, which depends on both the number of trials (n) and the probability of success (p). As the probability of success varies, so does the variance.

2. **Poisson Distribution**:
   - Mean (λ) = Variance (σ^2) = λ, where λ is the average rate of success (also known as the rate parameter).

   In the Poisson distribution, the mean and variance are equal and are both determined by the single parameter λ, which represents the average rate at which events occur in a fixed interval of time or space. This equality of mean and variance is a unique property of the Poisson distribution.

#  Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

In a normal distribution, also known as a Gaussian distribution, the least frequent data points appear at the tails of the distribution, furthest away from the mean. Specifically, the least frequent data points are located in the regions known as the tails or tails of the distribution curve.

In a standard normal distribution (with a mean of 0 and a standard deviation of 1), the tails are located at the extremes of the distribution, which are infinitely far from the mean in both positive and negative directions.

In a normal distribution with a non-zero mean, the least frequent data points still appear in the tails, but the location of the tails is shifted along the x-axis according to the mean of the distribution.
